<h1>Gráficos</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../../DataSets/netflix_titles.csv')
df = df.drop(df.columns[0], axis=1)
```

```python
df
```

<h2>Tratamento de dados do campo listed_in</h2>

```python
# Função para pegar o primeiro gênero antes da vírgula
def primeiro_gen(linha):
    linha['listed_in'] = linha['listed_in'].split(',')[0].strip()
    return linha
    

df = df.apply(lambda x: primeiro_gen(x), axis=1)


df
```

<h4>Contar quantos generos temos</h4>

```python
df['listed_in'].value_counts()
```

<h4>Pegando o eixo X com .keys que cria uma lista</h4>

```python
eixo_x = df['listed_in'].value_counts().keys()
print(eixo_x)
```

<h4>Pegando o eixo Y com .values que cria uma lista</h4>

```python
eixo_y = df['listed_in'].value_counts().values
print(eixo_y)
```

<h4>Colocando no gráfico de barras</h4>

<h4>Costumizando o gráfico</h4>

```python
plt.bar(eixo_x, eixo_y)
plt.title("Filmes por Gênero")
plt.xticks(rotation=90)
plt.xlabel("Nome categorias")
plt.ylabel("Quantidade de filmes")
#set parameters for tick labels
plt.tick_params(axis='x', which='major', labelsize=6)

plt.show()
```

<h4>Nova analise de dados, quantos sairam por ano</h4>

```python
df.head(3)
```

```python
# Titulo dos valores
eixo_x = df["release_year"].value_counts().sort_index().keys()
print(eixo_x)
# quantidade dos valores
eixo_y = df["release_year"].value_counts().sort_index().values
print(eixo_y)
```

```python
plt.plot(eixo_x,eixo_y)
plt.title("Lançamentos por anos")
plt.xlabel("Ano dos lançamentos")
plt.ylabel("Quantidade de filmes lançados")
```
